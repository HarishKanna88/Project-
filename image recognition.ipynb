```python
for i in range(len_of_images_tony):
    tony_gray[i]=np.ndarray.flatten(tony_gray[i]).reshape(flatten_size_tony,1)
```

```python
actor_1=pd.concat([edward_data,sharuk_data])
actor=pd.concat([actor_1,tony_data])
actor
```

```python
from sklearn.utils import shuffle
kollywood_indexed=shuffle(actor).reset_index()
kollywood_indexed
```

```python
kollywood_actors=kollywood_indexed.drop(['index'],axis=1)
kollywood_actors
```

```python
kollywood_actors.to_csv("C:/Users/91979/OneDrive/Desktop/edward/Kollywood_actors.csv")
```

```python
x=kollywood_actors.values[:,:-1]
y=kollywood_actors.values[:,-1]
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.7,random_state=0)
```

```python
from sklearn import svm
```

```python
clf=svm.SVC()
clf.fit(x_train,y_train)
```

```python
y_pred =clf.predict(x_test)
y_pred
```

```python
for i in range (25):
    predicted_images = (np.reshape(x_test[i],(512,512)).astype(np.float64))
    plt.title('predicted label:{0}'.format(y_pred[i]))
    plt.imshow(predicted_images,interpolation='nearest',cmap='gray')
    plt.show()
```

```python
from sklearn import metrics
accuracy=metrics.accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```
